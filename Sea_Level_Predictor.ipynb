```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ler o csv
df = pd.read_csv('/content/drive/MyDrive/flat-ui__data-Fri Jul 12 2024.csv')

# Verificar o head
print(df.head())
```

```python
# Criando o scatter plot
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])

# Criando a primeira linha do best fit
slope, intercept = np.polyfit(df['Year'], df['CSIRO Adjusted Sea Level'], 1)
years_extended = np.arange(df['Year'].min(), 2051)
line_of_best_fit = slope * years_extended + intercept
plt.plot(years_extended, line_of_best_fit, 'r', label='Best Fit Line (All Data)')

# Criando a segunda linha do best fit
df_recent = df[df['Year'] >= 2000]
slope_recent, intercept_recent = np.polyfit(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'], 1)
years_recent = np.arange(2000, 2051)
line_of_best_fit_recent = slope_recent * years_recent + intercept_recent
plt.plot(years_recent, line_of_best_fit_recent, 'green', label='Best Fit Line (Since 2000)')

# Add titulo e rotulos
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.legend()

# Salvar o plot e mostrar
plt.savefig('sea_level_plot.png')
plt.show()

```
